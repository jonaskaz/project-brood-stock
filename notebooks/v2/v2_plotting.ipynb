# Version 2 plots
The below plots were used to evaluate the performance of our system during long term testing. Plots that show expected brightness used log data from the microcontroller whereas plots that show actual brightness were taken using the ThorLabs powermeter PM100USB to measure actual light brightness.

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_csv('data/30MinuteCycle.csv')
df["hour_index"] = df.index /3600
sns.set_theme(style="darkgrid")
```

```python
sns.relplot(df, x="hour_index", y="brightness", kind="line", aspect=2).set(title="Manual Sunrise Sunset Cycle", ylabel="Expected Brightness Percentage", xlabel="hour")
```

The plot above shows the the expected brightness logged via serial for a ~1 hour cycle. Sunrise and sunset was set to 30 minutes in length. This plot shows the desired behavior of slowly ramping up, holding brightness until sunset, and slowly ramping down.

```python
df2 = pd.read_csv('data/OvernightSunrise.csv')
df2["hour_index"] = df2.index/60
sns.relplot(df2, x="hour_index", y="brightness", kind="line", aspect=2).set(title="Overnight Sunrise Test", ylabel="Expected Brightness Percentage", xlabel="Hour")
```

This test was run overnight using calculated sunrise and sunset times. This test began at midnight and ran until a bit after sunrise. The sunrise was set to 60 minutes in length. This shows expected behavior of staying off overnight, slowly ramping up for sunrise. and holding brightness afterwards.
